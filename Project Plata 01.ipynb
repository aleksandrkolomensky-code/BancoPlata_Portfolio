```python
import os
import pandas as pd
import random
from faker import Faker

# 1. Настройка путей для сохранения на Google Диск
PROJECT_DIR = '/content/drive/MyDrive/BancoPlata_Portfolio'
if not os.path.exists(PROJECT_DIR):
    os.makedirs(PROJECT_DIR)
    print(f"📁 Создана новая папка на Диске: {PROJECT_DIR}")
else:
    print(f"📁 Папка для проекта уже существует: {PROJECT_DIR}")

# 2. Инициализация генератора данных
fake = Faker()
Faker.seed(42)
random.seed(42)

roles = ['Frontend Developer', 'Backend Developer', 'QA Engineer', 'DevOps Engineer', 'Product Manager', 'Scrum Master']
squads = ['Squad Alpha', 'Squad Beta', 'Squad Gamma', 'Squad Delta']

role_rates = {
    'Frontend Developer': [40, 55],
    'Backend Developer': [45, 65],
    'QA Engineer': [30, 42],
    'DevOps Engineer': [50, 70],
    'Product Manager': [55, 75],
    'Scrum Master': [45, 60]
}

def generate_users(num_users=30):
    users_data = []
    for user_id in range(1, num_users + 1):
        name = fake.name()
        role = random.choice(roles)
        squad = random.choice(squads)

        rate_range = role_rates[role]
        cost_rate = random.randint(rate_range[0], rate_range[1])

        users_data.append({
            'user_id': user_id,
            'name': name,
            'role': role,
            'squad': squad,
            'cost_rate_usd': cost_rate
        })
    return pd.DataFrame(users_data)

# 3. Генерация и сохранение
df_users = generate_users(30)

# Путь к файлу на Диске
users_file_path = os.path.join(PROJECT_DIR, 'users.csv')
df_users.to_csv(users_file_path, index=False)

print(f"✅ Таблица 'users.csv' успешно создана и сохранена на Диск!")
print(f"📍 Путь к файлу: {users_file_path}")

# Посмотрим, что получилось
df_users.head()
```

```python
import os
import pandas as pd
import random

# Путь к нашей папке на Диске
PROJECT_DIR = '/content/drive/MyDrive/BancoPlata_Portfolio'

# Списки для генерации параметров задач
issue_types = ['Epic', 'Story', 'Tech Debt', 'Bug']
priorities = ['Low', 'Medium', 'High', 'Critical']
fibonacci_points = [1, 2, 3, 5, 8]

# Разделяем пользователей по ролям для реалистичного назначения
managers = df_users[df_users['role'].isin(['Product Manager', 'Scrum Master'])]['user_id'].tolist()
devs_and_qas = df_users[df_users['role'].isin(['Frontend Developer', 'Backend Developer', 'QA Engineer', 'DevOps Engineer'])]['user_id'].tolist()
all_user_ids = df_users['user_id'].tolist()

def generate_issues(num_issues=300):
    issues_data = []

    for issue_id in range(1, num_issues + 1):
        issue_key = f"BP-{1000 + issue_id}" # Banco Plata ID
        issue_type = random.choice(issue_types)
        priority = random.choice(priorities)

        # Логика для Story Points
        if issue_type in ['Story', 'Tech Debt']:
            story_points = random.choice(fibonacci_points)
        elif issue_type == 'Epic':
            story_points = 0 # Эпики обычно не оцениваются в SP напрямую
        else:
            story_points = 0 # У багов SP нет
            priority = random.choice(['High', 'Critical']) # Баги обычно приоритетнее

        # Реалистичное распределение авторов и исполнителей
        if issue_type == 'Epic':
            reporter = random.choice(managers)
            assignee = reporter # Менеджер сам ведет эпик
        else:
            reporter = random.choice(all_user_ids)
            assignee = random.choice(devs_and_qas)

        issues_data.append({
            'issue_id': issue_id,
            'issue_key': issue_key,
            'issue_type': issue_type,
            'priority': priority,
            'story_points': story_points,
            'reporter_id': reporter,
            'assignee_id': assignee
        })

    return pd.DataFrame(issues_data)

# Генерация задач
df_issues = generate_issues(300)

# Сохранение на Google Диск
issues_file_path = os.path.join(PROJECT_DIR, 'issues.csv')
df_issues.to_csv(issues_file_path, index=False)

print(f"✅ Таблица 'issues.csv' успешно создана и сохранена на Диск!")
print(f"📍 Путь к файлу: {issues_file_path}")

# Вывод для проверки
df_issues.head()
```

```python
import os
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import random

PROJECT_DIR = '/content/drive/MyDrive/BancoPlata_Portfolio'

np.random.seed(42)
random.seed(42)

def generate_perfect_changelogs(df_issues):
    changelog_data = []
    log_id = 1

    # Стартовая дата - весна 2026 года
    start_date = datetime(2026, 3, 1, 9, 0, 0)

    for idx, row in df_issues.iterrows():
        issue_id = row['issue_id']
        issue_type = row['issue_type']

        if issue_type == 'Epic':
            continue

        # Полный конвейер переходов с учетом времени ожидания ревью
        # (from_status, to_status, базовые_часы_ожидания)
        workflow_stages = [
            ('Backlog', 'Analysis', 12 if issue_type != 'Bug' else 2),
            ('Analysis', 'Ready for Dev', 24 if issue_type != 'Bug' else 1),
            ('Ready for Dev', 'In Progress', 48 if issue_type != 'Bug' else 2),
            ('In Progress', 'Ready for Code Review', 24 if issue_type != 'Bug' else 4),
            ('Ready for Code Review', 'Code Review', 16 if issue_type != 'Bug' else 2), # Время "зависания" в очереди на ревью
            ('Code Review', 'Ready for QA', 8 if issue_type != 'Bug' else 1),
            ('Ready for QA', 'QA In Progress', 16 if issue_type != 'Bug' else 4),
            ('QA In Progress', 'Ready for Release', 12 if issue_type != 'Bug' else 2),
            ('Ready for Release', 'Done', 24 if issue_type != 'Bug' else 3)
        ]

        # Время создания задачи
        current_time = start_date + timedelta(days=random.randint(0, 30), hours=random.randint(0, 8))

        for from_status, to_status, base_hours in workflow_stages:
            mean_log = np.log(base_hours)
            # Увеличим разброс (sigma) для очереди ревью, чтобы сымитировать "токсичные" задержки
            sigma_log = 1.1 if from_status == 'Ready for Code Review' else (0.8 if issue_type == 'Story' else 0.6)

            hours_spent = np.random.lognormal(mean_log, sigma_log)
            hours_spent = round(hours_spent, 2)

            transition_time = current_time + timedelta(hours=hours_spent)

            changelog_data.append({
                'log_id': log_id,
                'issue_id': issue_id,
                'from_status': from_status,
                'to_status': to_status,
                'changed_at': current_time.strftime('%Y-%m-%d %H:%M:%S'),
                'hours_spent': hours_spent
            })

            log_id += 1
            current_time = transition_time

    return pd.DataFrame(changelog_data)

# Генерация идеального журнала переходов
df_changelogs = generate_perfect_changelogs(df_issues)

# Сохранение на Google Диск
changelogs_file_path = os.path.join(PROJECT_DIR, 'changelogs.csv')
df_changelogs.to_csv(changelogs_file_path, index=False)

print(f"✅ Финальная таблица 'changelogs.csv' успешно сохранена на Диск!")
print(f"📍 Путь к файлу: {changelogs_file_path}")
print(f"📊 Сгенерировано записей о переходах: {len(df_changelogs)}")

df_changelogs.head(11)
```

```python
import os
import pandas as pd
from datetime import datetime

PROJECT_DIR = '/content/drive/MyDrive/BancoPlata_Portfolio'

# Загружаем то, что у тебя уже точно есть на Диске
df_issues = pd.read_csv(os.path.join(PROJECT_DIR, 'issues.csv'))
df_changelogs = pd.read_csv(os.path.join(PROJECT_DIR, 'changelogs.csv'))

def generate_worklogs(df_changelogs, df_issues):
    worklog_data = []
    worklog_id = 1
    issue_assignees = dict(zip(df_issues['issue_id'], df_issues['assignee_id']))

    for idx, row in df_changelogs.iterrows():
        issue_id = row['issue_id']
        from_status = row['from_status']
        hours_spent = row['hours_spent']
        changed_at_str = row['changed_at']

        user_id = issue_assignees.get(issue_id)
        if not user_id:
            continue

        start_time = datetime.strptime(changed_at_str, '%Y-%m-%d %H:%M:%S')

        # Нарезаем часы: в активных статусах инженеры пишут больше времени, в очередях - меньше
        if from_status in ['In Progress', 'Code Review', 'QA In Progress']:
            logged_hours = round(hours_spent * 0.75, 2)
        else:
            logged_hours = round(hours_spent * 0.10, 2)

        if logged_hours < 0.5:
            continue

        worklog_data.append({
            'worklog_id': worklog_id,
            'issue_id': issue_id,
            'user_id': user_id,
            'logged_hours': logged_hours,
            'logged_date': start_time.strftime('%Y-%m-%d')
        })
        worklog_id += 1

    return pd.DataFrame(worklog_data)

# Создаем и сохраняем недостающий файл
df_worklogs = generate_worklogs(df_changelogs, df_issues)
worklogs_file_path = os.path.join(PROJECT_DIR, 'worklogs.csv')
df_worklogs.to_csv(worklogs_file_path, index=False)

print(f"✅ Файл 'worklogs.csv' успешно создан и добавлен на твой Диск!")
```

```python
# Пишем аналитический SQL-запрос для расчета Эффективности потока
query_flow_efficiency = """
WITH status_categories AS (
    SELECT
        ch.issue_id,
        i.issue_type,
        u.squad,
        ch.from_status,
        ch.hours_spent,
        -- Классифицируем статусы на активную работу и ожидания
        CASE
            WHEN ch.from_status IN ('Analysis', 'In Progress', 'Code Review', 'QA In Progress') THEN 'Active'
            ELSE 'Wait'
        END as time_category
    FROM changelogs ch
    JOIN issues i ON ch.issue_id = i.issue_id
    JOIN users u ON i.assignee_id = u.user_id
),
aggregated_hours AS (
    SELECT
        squad,
        SUM(CASE WHEN time_category = 'Active' THEN hours_spent ELSE 0 END) as total_active_hours,
        SUM(CASE WHEN time_category = 'Wait' THEN hours_spent ELSE 0 END) as total_wait_hours
    FROM status_categories
    GROUP BY squad
)
SELECT
    squad as "Команда",
    ROUND(total_active_hours, 1) as "Активные часы (Active)",
    ROUND(total_wait_hours, 1) as "Часы ожидания (Wait)",
    ROUND((total_active_hours / (total_active_hours + total_wait_hours)) * 100, 2) || '%' as "Эффективность потока (Flow Efficiency)"
FROM aggregated_hours
ORDER BY total_active_hours / (total_active_hours + total_wait_hours) DESC;
"""

# Выполняем запрос и сохраняем результат в Pandas DataFrame
df_flow_results = pd.read_sql_query(query_flow_efficiency, conn)
df_flow_results
```

```python
# SQL-запрос для поиска конкретных узких мест в очередях
query_bottlenecks = """
SELECT
    from_status as "Статус ожидания",
    COUNT(*) as "Кол-во переходов",
    ROUND(AVG(hours_spent), 1) as "Среднее время (ч)",
    -- В SQLite нет встроенной медианы, но среднее арифметическое на наших логнормальных данных
    -- отлично подсветит "проблемные" зоны с длинными хвостами задержек
    ROUND(SUM(hours_spent), 1) as "Суммарный простой (ч)"
FROM changelogs ch
JOIN issues i ON ch.issue_id = i.issue_id
WHERE from_status IN ('Ready for Dev', 'Ready for Code Review', 'Ready for QA', 'Ready for Release')
GROUP BY from_status
ORDER BY SUM(hours_spent) DESC;
"""

df_bottlenecks = pd.read_sql_query(query_bottlenecks, conn)
df_bottlenecks
```

```python
import os

PROJECT_DIR = '/content/drive/MyDrive/BancoPlata_Portfolio'
app_code_path = os.path.join(PROJECT_DIR, 'app.py')

streamlit_code = """
import streamlit as st
import pandas as pd
import numpy as np
import plotly.express as px

st.set_page_config(page_title="Banco Plata - Engineering Operations Dashboard", layout="wide")

st.title("📊 Платформа анализа инженерной эффективности конвейера поставки")
st.markdown("---")

# 1. Загрузка данных
@st.cache_data
def load_data():
    users = pd.read_csv('users.csv')
    issues = pd.read_csv('issues.csv')
    changelogs = pd.read_csv('changelogs.csv')
    return users, issues, changelogs

try:
    df_users, df_issues, df_changelogs = load_data()
    # Связываем данные
    df_merged = df_changelogs.merge(df_issues, on='issue_id', how='inner')
    df_merged = df_merged.merge(df_users, left_on='assignee_id', right_on='user_id', how='inner')

    # ИНТЕРФЕЙС: Боковая панель фильтров
    st.sidebar.header("🎯 Фильтры конвейера")
    selected_squad = st.sidebar.multiselect(
        "Выбор Команды (Squad):",
        options=df_users['squad'].unique(),
        default=df_users['squad'].unique()
    )

    selected_type = st.sidebar.multiselect(
        "Тип задачи (Issue Type):",
        options=df_issues['issue_type'].unique(),
        default=['Story', 'Bug', 'Tech Debt']
    )

    # Фильтрация датасета
    df_filtered = df_merged[
        (df_merged['squad'].isin(selected_squad)) &
        (df_merged['issue_type'].isin(selected_type))
    ]

    # РАСЧЕТЫ: Flow Efficiency
    active_statuses = ['Analysis', 'In Progress', 'Code Review', 'QA In Progress']

    total_active = df_filtered[df_filtered['from_status'].isin(active_statuses)]['hours_spent'].sum()
    total_wait = df_filtered[~df_filtered['from_status'].isin(active_statuses)]['hours_spent'].sum()

    flow_efficiency = (total_active / (total_active + total_wait)) * 100 if (total_active + total_wait) > 0 else 0

    # ВЕРХНИЕ KPI КАРТОЧКИ
    col1, col2, col3 = st.columns(3)
    col1.metric("Эффективность потока (Flow Efficiency)", f"{flow_efficiency:.2f}%")
    col2.metric("Активное время работы (Active Time)", f"{total_active:,.1f} ч")
    col3.metric("Время простоев в очередях (Wait Time)", f"{total_wait:,.1f} ч")

    st.markdown("---")

    # СЕКЦИЯ ГРАФИКОВ
    col_chart1, col_chart2 = st.columns(2)

    with col_chart1:
        st.subheader("⏳ Распределение времени ожидания по очередям")
        wait_stages = ['Ready for Dev', 'Ready for Code Review', 'Ready for QA', 'Ready for Release']
        df_wait = df_filtered[df_filtered['from_status'].isin(wait_stages)]
        df_wait_grouped = df_wait.groupby('from_status')['hours_spent'].sum().reset_index()

        fig_bar = px.bar(
            df_wait_grouped,
            x='from_status',
            y='hours_spent',
            labels={'from_status': 'Статус очереди', 'hours_spent': 'Суммарный простой (часы)'},
            color='from_status',
            color_discrete_sequence=px.colors.sequential.Oranges_r
        )
        st.plotly_chart(fig_bar, use_container_width=True)

    with col_chart2:
        st.subheader("👥 Сравнение эффективности команд")
        squad_data = []
        for squad in selected_squad:
            df_sq = df_filtered[df_filtered['squad'] == squad]
            act = df_sq[df_sq['from_status'].isin(active_statuses)]['hours_spent'].sum()
            wt = df_sq[~df_sq['from_status'].isin(active_statuses)]['hours_spent'].sum()
            eff = (act / (act + wt)) * 100 if (act + wt) > 0 else 0
            squad_data.append({"Команда": squad, "Эффективность (%)": round(eff, 2)})

        fig_squad = px.bar(
            pd.DataFrame(squad_data),
            x='Команда',
            y='Эффективность (%)',
            range_y=[0, 100],
            color='Эффективность (%)',
            color_continuous_scale=px.colors.sequential.Viridis
        )
        st.plotly_chart(fig_squad, use_container_width=True)

except Exception as e:
    st.error(f"Не удалось загрузить данные. Проверьте файлы в корневой папке. Ошибка: {e}")
"""

with open(app_code_path, 'w', encoding='utf-8') as f:
    f.write(streamlit_code)

print(f"🎉 Код дашборда успешно сгенерирован и сохранен в app.py!")
print(f"📍 Проверь папку на Диске, там появился новый файл!")
```
